```python
import yfinance as yf
import pandas as pd
from datetime import datetime
import pytz
list_symbols = pd.read_excel('/Users/alexandre/Desktop/FIIs/FiisPY/carteira.xlsx')['CÃ³digo'].values
list_symbols
```

```python
ativos = []
list_precos = []
dias = []
for ativo in list_symbols:
    dados = yf.Ticker(ativo+'.SA')
    hist = dados.history(period="12mo")
    hist_precos = hist['Close']
    hist_precos_ultimo_ano = hist_precos.loc[datetime(2022,10,1, tzinfo=pytz.utc):datetime(2023,10,30, tzinfo=pytz.utc)]
    total12 = 0
    for media1 in hist_precos:
        total12 += media1   
    list_precos.append(total12/len(hist_precos_ultimo_ano))
    ativos.append(ativo)
```

```python
carteira_df = pd.DataFrame({'Ativos':ativos, 'Valor de Compra':list_precos}).sort_values(by = 'Valor de Compra',ascending=True)
carteira_df
```
